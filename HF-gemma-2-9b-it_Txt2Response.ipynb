```python
from transformers import pipeline
import torch

model_id = "google/gemma-2-9b-it"

pipe = pipeline(
    "text-generation",
    model = model_id,
    model_kwargs = {"torch_dtype" : torch.bfloat16},
    device_map = "auto"
)
```

```python
messages = [
    {"role": "user", "content": "write 10 line about generative ai "},
    ]

output = pipe(
    messages,
    max_new_tokens=256,
    do_sample=False,
)

assistant_response = output[0]["generated_text"][-1]["content"]
print(assistant_response)
```
